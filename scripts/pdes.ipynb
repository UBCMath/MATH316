```python
import numpy as np

import pdes
VIEW = (30, 180 + 45, 0)
```

```python
IC = lambda x: np.cos(np.pi * x / 2)

x, t, u = pdes.diffusion_1d(1, 'D', 'D', 0, 0, IC, 1, [0, 0.5], 51, 100001)
pdes.plot_2d(x, u)
```

```python
F = lambda x: np.cos(3 * np.pi * x)
G = lambda x: np.cos(np.pi * x)

x, t, u = pdes.wave_1d(
    0.5, 
    'N', 'N', 0, 0, 
    F, G, 
    1, [0, 6], 101, 8001
)

pdes.plot_2d(x, u)
```

```python
pdes.plot_3d(x, t, u, VIEW)
```

```python
L = lambda x,y: 1
R = lambda x,y: (2 - y) * y
T = lambda x,y: 0
B = lambda x,y: 0

x, y, u, k = pdes.laplace_2d(
    'N', 'D', 'D', 'D',
    L, R, T, B,
    3, 2, 61, 41,
    1E-3
)

print(k)
pdes.plot_3d(x, y, u, VIEW)
```
